```python
len(usaBusiness)
```

```python
usaBusiness=usaBusiness.dropna(subset=["categories"])
len(usaBusiness)
```

```python
usaRestuarants=usaBusiness[usaBusiness['categories'].str.contains('Restaurants','Food')]
len(usaRestuarants)
```

```python
usa_not_restuarants = usaBusiness[~usaBusiness['categories'].str.contains('Restaurants')]
usa_not_restuarants
```

```python
CategoriesSet = []

for row in usa_not_restuarants["categories"]:
    for i in row.split(","):
        CategoriesSet.append(i)
```
